```python
def calculate_max_drawdown(prices):

    # Calculate the cumulative maximum
    cumulative_max = prices.cummax()
    
    # Calculate the drawdown
    drawdown = (prices - cumulative_max) / cumulative_max
    
    return np.min(drawdown)
```
